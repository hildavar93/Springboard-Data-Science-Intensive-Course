## What is Citi Bike?
Citi Bike is "the nation's largest bike share program" with approximately 10,000 bikes and over 600 stations in select areas of New York City and Jersey City. If you have traveled to New York City since Citi Bike's inception in 2013, you have probably noticed the rows of blue bicycles that are seemingly located on every other street corner in Manhattan. Marketed as an affordable, eco-friendly, and healthy alternative to New York City's subway system and ride-sharing apps such as Uber, Citi Bike has become an increasingly popular transportation method in the last three years.

Citi Bike generously provides some of their data on its website https://www.citibikenyc.com/system-data. As of December 2016, there have been 36,902,025 Citi Bike rides.

## Who Uses Citi Bike?
The only data that Citi Bike provides concerning the demographics of its user base are users' genders and birth years. Because the data provided does not include a unique identifier per user, it is difficult to say how many female and male users use Citi Bike. However, we can calculate the number of rides taken by men and women at given times. We can also calculate the number of rides taken by people in different ages at given times.

According to the graph of the number of trips vs. age range, most of Citi Bike users are between the age range of 1966-1995. The overwhelming majority of trips, however, were taken by people born between 1976-1985.

I was quite surprised by the large disparity between rides taken by men and those taken by women. According to my barplot below, trips taken by men account for about 75% of all of Citi Bike's rides since its inception. There is a good chance that most of Citi Bike's regular users are male, but there might be more that Citi Bike could be doing to market towards women.

One trend that could be interesting to explore further is how ridership changes over time with respect to gender and age group.

```python
import numpy as np
import pandas as pd

# Null values written as "/N" -- replace with nulls
peopledf=pd.read_csv("/Users/hildavargas/Desktop/Springboard_Mini_Projects/Capstone_Project/CB_real_metrics/PeopleGenderAge.csv", na_values=[r'\N'])

# Replace Gender w/ 'Unknown, Male, Female' vs. '0, 1, 2'
peopledf.Gender=peopledf.Gender.replace(0, 'Unknown')
peopledf.Gender=peopledf.Gender.replace(1, 'Male')
peopledf.Gender=peopledf.Gender.replace(2, 'Female')

```

```python
total_trips=np.sum(peopledf['Number of Trips'])
total_trips
```

```python
x=peopledf.copy()
z=x['Number of Trips'].groupby([x.Year]).sum().reset_index()
z.head()
```

```python
mon=x['Number of Trips'].groupby([x.Month]).sum().reset_index()
mon.head()
```

```python
# Replace birth years < 1920 with nulls (see notebook "Identifying Outliers in Age Range" for reference)
peopledf.ix[peopledf['Birth Year'] < 1920, 'Birth Year'] = np.nan

# Drop rows with nulls (missing birth years and those < 1920)
peopledf=peopledf.dropna()
peopledf.head()
```

```python
# Create bins for age ranges based off the birth year values
bins=[1915, 1925, 1935, 1945, 1955, 1965, 1975, 1985, 1995, 2005]

# Add age ranges column to dataframe
peopledf['Age Ranges']=pd.cut(peopledf['Birth Year'], precision=0, bins=bins)
del peopledf['Birth Year']
peopledf.sort_values('Age Ranges')
peopledf.head()
```

```python
# Checking for nulls in age range column
x=peopledf['Age Ranges'].isnull()
np.sum(x)
```

```python
# Change age ranges to intervals that will be easier to understand in a plot
unique_ranges=peopledf['Age Ranges'].unique()
from collections import defaultdict
x=defaultdict(str)

for line in unique_ranges:
    z=line[1:4]+'6'
    d=line[7:11]
    result=z+'-'+d
    x[line]=result

# Update dataframe to reflect new intervals
peopledf['Age Range']=peopledf['Age Ranges'].map(x)
peopledf.head()
```
